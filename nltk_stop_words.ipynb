### 📘 Stop Words in NLP (Lecture Notes)

✅ 1. What are Stop Words?

Stop words are common words that usually do not carry significant meaning in text processing.
Examples:

    is ,    the ,    am ,    are
    in ,    on ,    and,     of

👉 These words are often removed during text preprocessing.


✅ 2. Why Remove Stop Words?

✔ Reduces text size
✔ Improves model performance
✔ Removes noise
✔ Speeds up processing

Example:

Original:

    "NLP is very important in the field of Artificial Intelligence"

After removing stopwords:

    "NLP important field Artificial Intelligence"

✅ Example 2: Remove Stop Words Using spaCy
python

import spacy
nlp = spacy.load("en_core_web_sm")

text = "Natural Language Processing is very important in the modern world."

doc = nlp(text)

# Remove stop words
filtered_words = []

for token in doc:
    if not token.is_stop:  # Keep only non-stopwords
        filtered_words.append(token.text)

print("After removing stopwords:", filtered_words)

✅ Example 3: Stop Words Using NLTK
python

import nltk
nltk.download('stopwords')
nltk.download('punkt')

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

text = "This is an example showing stop word removal using NLTK."

# Tokenize text
words = word_tokenize(text)

# Load English stopwords
stop_words = set(stopwords.words("english"))

# Remove stopwords
filtered = []

for word in words:
    if word.lower() not in stop_words:
        filtered.append(word)

print("Filtered words:", filtered)

```python
paragraph = ('Natural Language Processing (NLP) is supported by several powerful libraries in Python, each designed for different purposes. NLTK (Natural Language Toolkit) is one of the oldest and most widely used libraries, mainly focused on education and research; it provides tools for tokenization, stemming, lemmatization, parsing, and access to many linguistic corpora. spaCy is a modern, fast, and production-oriented NLP library that follows an object-oriented approach, returning Doc, Token, and Span objects, and supports tasks such as part-of-speech tagging, named entity recognition (NER), dependency parsing, and word vectors. TextBlob is built on top of NLTK and Pattern, offering a simple interface for beginners to perform tasks like sentiment analysis, translation, noun phrase extraction, and text classification. Gensim is mainly used for topic modeling and vector space modeling, including Word2Vec, Doc2Vec, and LDA. Stanford NLP (Stanza) provides advanced deep learning–based linguistic analysis with high accuracy and supports multiple languages. Transformers (Hugging Face) is a powerful library that provides pre-trained state-of-the-art deep learning models like BERT, GPT, and RoBERTa for tasks such as text classification, question answering, summarization, and translation. Additionally, Flair is known for contextual string embeddings, and FastText by Facebook is efficient for word embeddings and text classification. Together, these libraries enable researchers and developers to perform rule-based, statistical, and deep learning–based NLP tasks efficiently.')
```

```python
from nltk.stem import PorterStemmer
stemmer=PorterStemmer()
sentences = nltk.sent_tokenize(paragraph)
```

```python
type(sentences)
```

```python
## Apply stopwords and Filter and apply Stemmer
for i in range(len(sentences)):
    words=nltk.word_tokenize(sentences[i])
    words = [stemmer.stem(word) for word in words if word not in set(stopwords.words('english'))] # list compression
    sentences[i] = ''.join(words)# converting all the list of  words into sentences
```

```python
from nltk.stem import SnowballStemmer
snowballstemmer=SnowballStemmer('english')
```

```python
## Apply stopwords and Filter and apply snowball Stemmer
for i in range(len(sentences)):
    words=nltk.word_tokenize(sentences[i])
    words = [snowballstemmer.stem(word) for word in words if word not in set(stopwords.words('english'))] # list compression
    sentences[i] = ''.join(words)# converting all the list of  words into sentences
```

```python
sentences
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
## Apply stopwords and Filter and apply snowball Stemmer
for i in range(len(sentences)):
    words=nltk.word_tokenize(sentences[i])
    words = [lemmatizer.lemmatize(word, pos='v') for word in words if word not in set(stopwords.words('english'))] # list compression
    sentences[i] = ''.join(words)# converting all the list of  words into sentences
```
