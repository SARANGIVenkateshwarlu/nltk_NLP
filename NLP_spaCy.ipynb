### 📘 spaCy and Object-Oriented Approach (Lecture Notes)
✅ 1. Introduction to spaCy

    spaCy is a powerful NLP library in Python.
    It follows an object-oriented approach.
    When we process text, spaCy returns a Doc object.
    Inside the Doc:
        Each word is a Token object
        Each sentence is a Span object

```python


So the structure is:

Text → Doc → Tokens & Sentences (Span)

✅ 2. Install and Load spaCy
Install (run once in terminal):
bash

pip install spacy
python -m spacy download en_core_web_sm

✅ 3. Basic Example (Creating Doc Object)
python

import spacy

# Load English model
nlp = spacy.load("en_core_web_sm")

# Input text
text = "spaCy uses object-oriented approach. It is very powerful."

# Process text
doc = nlp(text)

print(type(doc))

✅ Output:

<class 'spacy.tokens.doc.Doc'>

✔ doc is an object of class Doc
✅ 4. Token Objects (Words as Objects)

Each word in doc is a Token object.
python

for token in doc:
    print(token.text, "---->", type(token))

✅ Output Example:
angelscript

spaCy ----> <class 'spacy.tokens.token.Token'>
uses ----> <class 'spacy.tokens.token.Token'>

✔ Each word is a Token object.
✅ 5. Accessing Token Properties

Each Token object has attributes:
python

for token in doc:
    print("Word:", token.text)
    print("Lemma:", token.lemma_)
    print("POS:", token.pos_)
    print("Is Stopword:", token.is_stop)
    print("-----")

✅ 6. Sentence Objects (Span Objects)

Sentences inside a Doc are Span objects.
python

for sentence in doc.sents:
    print(sentence)
    print(type(sentence))

✅ Output:

spaCy uses object-oriented approach.
<class 'spacy.tokens.span.Span'>

✔ Each sentence is a Span object.
✅ 7. Visual Structure

Doc Object
│
├── Token Object (word)
├── Token Object (word)
├── Span Object (sentence)

✅ 8. Test Code for Students
📝 Test 1: Identify Object Types
python

import spacy
nlp = spacy.load("en_core_web_sm")

doc = nlp("Natural Language Processing is interesting.")

print(type(doc))

for token in doc:
    print(token.text, type(token))

for sent in doc.sents:
    print(sent, type(sent))

📝 Test 2: Access Attributes
python

doc = nlp("Students are learning NLP with spaCy.")

for token in doc:
    print(token.text, token.pos_, token.lemma_)

✅ 9. Important Exam Points

✔ spaCy follows object-oriented design
✔ nlp() returns a Doc object
✔ Words → Token objects
✔ Sentences → Span objects
✔ Each object has properties and methods
✅ 10. Short Theory Answer (For Exams)

    spaCy uses an object-oriented approach where processed text is stored in a Doc object. Each word in the document is represented as a Token object, and each sentence is represented as a Span object. These objects contain linguistic attributes such as lemma, part-of-speech, and dependency information.
```
