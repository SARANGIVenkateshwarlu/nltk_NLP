NLTK

```python
import nltk
```

```python
corpus = """Natural Language Processing (NLP) plays an important role in enabling machines to understand and generate human language.
 Natural Language Toolkit (NLTK) stands out as one of the most widely used libraries.
   It provides a combination linguistic resources.
     Ii including text processing libraries and pre-trained models. 
Which makes it ideal for both academic research and practical applications.
"""
```

```python
print(corpus)
```

```python
### Tokentization
### Sentence to paragraph 
### text to split into sentences
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
# Download required resources (run once)
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
print(word_tokenize(corpus))
```

```python
print(sent_tokenize(corpus))
```

```python
documents= sent_tokenize(corpus)
```

```python
type(documents)
```

```python
for sentence in documents:
    print(sentence)
```

```python
words_list=word_tokenize(corpus)
type(words_list)
```

```python
for word in words_list:
    print(word)
```

```python
from nltk.tokenize import wordpunct_tokenize
```

```python
wordpunct_tokenize(corpus)
```

```python
from nltk.tokenize import TreebankWordTokenizer
```

```python
tokenizer = TreebankWordTokenizer()
tokenizer.tokenize(corpus)
```
