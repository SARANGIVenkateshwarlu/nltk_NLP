### What is Named Entity Recognition (NER)?

Named Entity Recognition (NER) is a task in Natural Language Processing (NLP) where we identify and classify:

    👤 PERSON names
    🏢 ORGANIZATIONS
    🌍 LOCATIONS
    📅 DATES
    💰 MONEY
    📊 PERCENTAGES
    and more


```python
sentence = 'Apple hired John Smith in California in 2024'
```

NER Output:

    Apple → ORGANIZATION
    John Smith → PERSON
    California → LOCATION
    2024 → DATE


🧩 How NLTK Does NER

NLTK uses:

    Tokenization → split sentence into words
    POS tagging → assign grammatical tags
    Chunking (NE chunker) → identify named entities

NLTK uses a pretrained model for English entity chunking.

```python
# Install Required Packages
#!pip install nltk
import nltk
```

```python
# download required resources:
import nltk

# Download required NLTK datasets
nltk.download('punkt')           # tokenizer
nltk.download('averaged_perceptron_tagger')  # POS tagger
nltk.download('maxent_ne_chunker')           # NER chunker
nltk.download('words')           # word corpus
```

```python
# Basic NER Example
import nltk
from nltk import word_tokenize, pos_tag, ne_chunk

# Sample text
text = "Apple hired John Smith in California in 2024."

# Step 1: Tokenize sentence into words
tokens = word_tokenize(text)

# Step 2: Perform Part-of-Speech tagging
pos_tags = pos_tag(tokens)

# Step 3: Perform Named Entity Recognition
named_entities = ne_chunk(pos_tags)

# Print result
print(named_entities)
```

### What Each Step Does

1. word_tokenize()

Breaks sentence into words. 
["Apple", "hired", "John", "Smith", "in", "California", "in", "2024", "."]

2. pos_tag()

Assigns grammatical role:
[('Apple', 'NNP'),
 ('hired', 'VBD'),
 ('John', 'NNP'),
 ('Smith', 'NNP'),
 ('California', 'NNP')]
 NNP = Proper Noun

3. ne_chunk()

Groups words into named entity trees.

Output looks like:
(S
  (ORGANIZATION Apple)
  hired
  (PERSON John Smith)
  in
  (GPE California)
  in
  2024
  .)

```python
#Extract Only Named Entities
#To extract clean entities:
def extract_entities(tree):
    entities = []
    
    for subtree in tree:
        # Check if subtree has a label (meaning it's an entity)
        if hasattr(subtree, 'label'):
            entity_name = " ".join([token for token, pos in subtree.leaves()])
            entity_type = subtree.label()
            entities.append((entity_name, entity_type))
    
    return entities


entities = extract_entities(named_entities)

print("Extracted Entities:")
print(entities)
```

Concept flow: Text → Tokens → POS Tags → Chunk Parser → Named Entities
